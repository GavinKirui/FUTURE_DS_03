```python
#Import Libraries for data manipulation and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("C:/Users/USER/Desktop/tmdb_5000_movies.csv")
```

```python

# Display the first 10 rows to understand the structure
data.head(10)
```

```python
# Check the shape of the dataset
print(f"Dataset contains {data.shape[0]} rows and {data.shape[1]} columns.")
```

```python
# View basic info about the dataset
data.info()
```

```python
# Check for missing values
data.isnull().sum()
```

```python
# Drop columns like 'homepage' and 'tagline' with too many missing values
data = data.drop(columns=['homepage', 'tagline'], errors='ignore')
```

```python
#Check if homepage and taglines columns are eliminated
print(data.columns)
```

```python
# Replace invalid 'release_date' values with NaT
data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')
```

```python
data.isnull().sum()
```

```python
# Display basic statistics for numerical columns
data.describe()

```

```python
#Univariate Analysis
#To see how the movie budgets are distributed
plt.figure(figsize=(10, 5))
sns.histplot(data['runtime'], bins=20, kde=True, color='blue')
plt.title("Distribution of Movie Runtimes")
plt.xlabel("Runtime (minutes)")
plt.ylabel("Frequency")
plt.show()
```

```python
#visualize the revenue distribution of the movies
plt.figure(figsize=(10, 8))
sns.histplot(data['revenue'], bins=20, kde=True, color='purple')
plt.title("Distribution of Movie Revenues")
plt.xlabel("Revenue (in millions)")
plt.ylabel("Frequency")
plt.show()
```

```python
#To see how the popularity of movies is distributed:
plt.figure(figsize=(10, 5))
sns.histplot(data['popularity'], bins=20, kde=True, color='orange')
plt.title("Distribution of Movie Popularity")
plt.xlabel("Popularity Score")
plt.ylabel("Frequency")
plt.show()
```

```python
#Vote Average Distribution
plt.figure(figsize=(10, 5))
sns.histplot(data['vote_average'], bins=20, kde=True, color='red')
plt.title("Distribution of Vote Averages")
plt.xlabel("Vote Average")
plt.ylabel("Frequency")
plt.show()
```

```python
# Extract the year
data['release_year'] = data['release_date'].dt.year

# Plotting the distribution of release years
plt.figure(figsize=(10, 5))
sns.histplot(data['release_year'], bins=30, kde=True, color='purple')
plt.title("Distribution of Movie Release Years")
plt.xlabel("Release Year")
plt.ylabel("Frequency")
plt.show()
```

```python
# Correlation heatmap
corr_matrix = data[['budget', 'revenue', 'popularity']].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation between Budget, Revenue, and Popularity")
plt.show()
```

```python
# Scatter plot: runtime vs vote_average
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='runtime', y='vote_average', color='blue', alpha=0.6)
plt.title("Runtime vs Vote Average")
plt.xlabel("Runtime (minutes)")
plt.ylabel("Vote Average")
plt.show()

```

```python
# Scatter plot: runtime vs revenue
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='runtime', y='revenue', color='green', alpha=0.6)
plt.title("Runtime vs Revenue")
plt.xlabel("Runtime (minutes)")
plt.ylabel("Revenue")
plt.show()
```

```python
#Analyze genre popularity or revenue trends
import ast  # For parsing genres stored as strings

# Parse genres into lists
data['genres_list'] = data['genres'].apply(lambda x: [g['name'] for g in ast.literal_eval(x)])

# Explode genres for groupby operations
genre_data = data.explode('genres_list')

# Calculate mean revenue and popularity for each genre
genre_summary = genre_data.groupby('genres_list')[['revenue', 'popularity']].mean().sort_values(by='revenue', ascending=False)
print(genre_summary)

# Visualize genre trends
genre_summary.plot(kind='bar', figsize=(12, 6), title="Genre Revenue and Popularity", ylabel="Average")
plt.show()
```

```python
# Parse production_countries into lists
data['countries_list'] = data['production_countries'].apply(lambda x: [c['name'] for c in ast.literal_eval(x)])

# Explode countries for groupby
country_data = data.explode('countries_list')

# Aggregate revenue and count of movies by country
country_summary = country_data.groupby('countries_list')['revenue'].agg(['sum', 'count']).sort_values(by='sum', ascending=False)
print(country_summary)

# Visualize top 10 countries by revenue
country_summary.head(10).plot(kind='bar', figsize=(12, 6), title="Top 10 Countries by Revenue", ylabel="Revenue")
plt.show()
```

```python
#Year-on-year changes in production budgets and revenues
# Group data by release year
yearly_data = data.groupby('release_year')[['budget', 'revenue']].mean()

# Plot trends over years
plt.figure(figsize=(12, 6))
plt.plot(yearly_data.index, yearly_data['budget'], label='Average Budget', marker='o')
plt.plot(yearly_data.index, yearly_data['revenue'], label='Average Revenue', marker='o')
plt.title("Year-on-Year Changes in Budget and Revenue")
plt.xlabel("Year")
plt.ylabel("USD")
plt.legend()
plt.grid(True)
plt.show()
```
